```python
#view missing data
missing_data_sorted = df.isnull().sum()[df.isnull().sum() > 0].sort_values(ascending=False)

print(missing_data_sorted)
```

```python
#impute missing data
df['PoolQC_imputed'] = df['PoolQC'].fillna('NA')
df['MiscFeature_imputed'] = df['MiscFeature'].fillna('NA')
df['Alley_imputed'] = df['Alley'].fillna('NA')
df['Fence_imputed'] = df['Fence'].fillna('NA')
df['FireplaceQu_imputed'] = df['FireplaceQu'].fillna('NA')
df['LotFrontage_imputed'] = df['LotFrontage'].fillna(0)
df['GarageType_imputed'] = df['GarageType'].fillna('NA')
df['GarageYrBlt_imputed'] = df['GarageYrBlt'].fillna(0)
df['GarageFinish_imputed'] = df['GarageFinish'].fillna('NA')
df['GarageQual_imputed'] = df['GarageQual'].fillna('NA')
df['GarageCond_imputed'] = df['GarageCond'].fillna('NA')
df['BsmtQual_imputed'] = df['BsmtQual'].fillna('NA')
df['BsmtCond_imputed'] = df['BsmtCond'].fillna('NA')
df['BsmtExposure_imputed'] = df['BsmtExposure'].fillna('NA')
df['BsmtFinType1_imputed'] = df['BsmtFinType1'].fillna('NA')
df['BsmtFinType2_imputed'] = df['BsmtFinType2'].fillna('NA')
df['MasVnrType_imputed'] = df['MasVnrType'].fillna('None')
df['MasVnrArea_imputed'] = df['MasVnrArea'].fillna(0)
df['Electrical_imputed'] = df['Electrical'].fillna('NA')
df_clean = df.dropna(axis=1, how='any')
missing_data_sorted = df_clean.isnull().sum()[df.isnull().sum() > 0].sort_values(ascending=False)

print(missing_data_sorted)
print(df_clean.head())
```

```python
#get list of disinct value sums for each nonnumeric column
distinct_counts_sorted = df_clean.select_dtypes(exclude=['number']).nunique().sort_values(ascending=False)
print(distinct_counts_sorted)


```

```python
#Lets get stupid

df_dumb = pd.get_dummies(df_clean)

print(df_dumb.head())
```

```python
#############################################FIXME#############################################
# Calculate the correlation matrix
corr_matrix = df_dumb.corr()

# Unstack the matrix to get pairs of correlations, sort by absolute value, and filter out self-correlations (correlation with itself)
sorted_corr = corr_matrix.abs().unstack().sort_values(ascending=False)

# Remove self-correlations (where a column correlates with itself)
sorted_corr = sorted_corr[sorted_corr < 1]

# Display the top 10 highest correlations
top_50_corr = sorted_corr.head(50)
print(top_50_corr)

```

```python
#Feature engineering time
#drop dependent vars
df_X = df_dumb.drop(['SalePrice'], axis=1)
df_clean = df_clean.drop(['SalePrice'], axis=1)
#Get list of numeric columns
numeric_columns = df_clean.select_dtypes(include=['number']).columns
#Get log vars
for col in numeric_columns:
    df_X[f'log_{col}'] = np.log1p(df_clean[col])  # Apply log1p transformation
```
